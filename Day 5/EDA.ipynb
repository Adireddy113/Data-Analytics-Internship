```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

df = pd.read_csv(r"C:\Users\Administrator\Downloads\titanic_sample.csv")
print(df.head())

print(df.info())

print("\nðŸ”¹ Summary Statistics:")
print(df.describe(include='all'))

print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())

sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()

sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()

sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()

sns.histplot(df['Age'].dropna(), kde=True, bins=20)
plt.title("Age Distribution")
plt.show()

sns.boxplot(x='Fare', data=df)
plt.title("Fare Boxplot (Check for outliers)")
plt.show()

sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.show()

sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Class")
plt.show()

sns.countplot(x='Survived', hue='Embarked', data=df)
plt.title("Survival by Embarkation Port")
plt.show()

sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare (Survival Color Coded)")
plt.show()

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()

df.drop(columns='Cabin', inplace=True)


print("\n Summary of EDA Findings:")
print("""
1. Women and passengers in 1st class had a much higher survival rate.
2. Most passengers were from 3rd class and embarked from 'S'.
3. Age and Fare have some outliers; Fare is right-skewed.
4. There's a positive correlation between Pclass and Fare.
5. Gender and class are major influencers of survival.
""")
```
